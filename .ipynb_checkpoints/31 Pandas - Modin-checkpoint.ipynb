```python
import time
import pandas as pd
df = pd.read_csv('test_31.csv')
s = time.time()
df = pd.concat([df for _ in range(100)])
e = time.time()
print("Pandas Concat Time = {}".format(e-s))
```

```python
import time
import modin.pandas as pd
df = pd.read_csv('test_31.csv')
s = time.time()
df = pd.concat([df for _ in range(100)])
e = time.time()
print("Modin Pandas Concat Time = {}".format(e-s))
```
